```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
income = pd.read_csv('D:\\Data Analysis and Mining\\data\\train.csv')
```

```python
income.apply(lambda x:np.sum(x.isnull()))
```

```python
income.head()
```

```python
income.fillna(value = {'workclass':income.workclass.mode()[0],
              'occupation':income.occupation.mode()[0],
              'native-country':income['native-country'].mode()[0]},inplace=True)
```

```python
income.describe()
```

```python
income.describe(include = ['object'])
```

```python
plt.style.use('ggplot')
fig,axes = plt.subplots(2,1)
plt.figure(figsize = (10,6))
income.age[income['income_>50K'] == 0].plot(kind='kde',label = '<=50K',ax = axes[0],
                                         legend = True,linestyle = '-')
income.age[income['income_>50K'] == 1].plot(kind='kde',label = '>50K',ax = axes[0],
                                         legend = True,linestyle = '--')

income['hours-per-week'][income['income_>50K'] == 0].plot(kind = 'kde',label = '<=50K',ax = axes[1],
                                                       legend = True,linestyle = '-')
income['hours-per-week'][income['income_>50K'] == 1].plot(kind = 'kde',label = '>50K',ax = axes[1],
                                                       legend = True,linestyle = '--')

plt.show()
```

```python
plt.close()
```

```python
race = pd.DataFrame(income.groupby(by = ['race','income_>50K']).aggregate(np.size).loc[:'age'])
race = race.reset_index()
race.rename(columns = {'age':'counts'},inplace = True)
race.sort_values(by = ['race','counts'],ascending = False,inplace = True)

relationship = pd.DataFrame(income.groupby(by = ['relationship','income_>50K']).aggregate(np.size).loc[:'age'])
relationship = relationship.reset_index()
relationship.rename(columns = {'age':'counts'},inplace = True)
relationship.sort_values(by = ['relationship','counts'],ascending = False,inplace = True)

plt.figure(figsize = (10,6))
sns.barplot(x = 'race',y = 'counts',hue = 'income_>50K',data = race)
plt.show()

plt.figure(figsize = (10,6))
sns.barplot(x = 'relationship',y = 'counts',hue = 'income_>50K',data = relationship)
plt.show()
```

```python
for feature in income.columns:
    if income[feature].dtype == 'object':
        income[feature] = pd.Categorical(income[feature]).codes

income.head()
```

```python
income.drop(['education','fnlwgt'],axis = 1,inplace = True)
income.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_valid,y_train,y_valid = train_test_split(income.loc[:,'age':'native-country'],income['income_>50K'],train_size = 0.8,random_state = 1234)
print('training dataset has {} pieces of data.'.format(x_train.shape[0]))
print('validing dataset has {} pieces of data.'.format(x_valid.shape[0]))
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(x_train,y_train)

print(kn.get_params())
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbdt = GradientBoostingClassifier()
gbdt.fit(x_train,y_train)

print(gbdt.get_params())
```

```python
from sklearn.model_selection import GridSearchCV

k_options = list(range(1,12))
parameters = {'n_neighbors':k_options}

grid_kn = GridSearchCV(estimator = KNeighborsClassifier(),param_grid = parameters,cv = 10,scoring = 'accuracy')
grid_kn.fit(x_train,y_train)
grid_kn.cv_results_
```

```python
grid_kn.score(x_train,y_train)
```

```python
parameters = {
    "loss":["deviance"],
    "learning_rate": [0.01,0.2],
    "min_samples_split": np.linspace(0.1, 0.3, 3),
    "min_samples_leaf": np.linspace(0.1, 0.3, 3),
    "max_depth":[3,5,8],
    "max_features":["sqrt"],
    "criterion": ["mae"],
    "subsample":[0.5,0.75,1.0],
    "n_estimators":[10]
    }
grid_gbdt = GridSearchCV(estimator = GradientBoostingClassifier(),param_grid = parameters,cv = 10,scoring = 'accuracy')
grid_gbdt.fit(x_train,y_train)
grid_gbdt.cv_results_
```

```python
grid_gbdt.score(x_train,y_train)
```

```python
kn_pred = kn.predict(x_valid)
print(pd.crosstab(kn_pred,y_valid))
print('kn model\' accuary on training dataset is {}'.format(kn.score(x_train,y_train)))
print('kn model\' accuary on validing dataset is {}'.format(kn.score(x_valid,y_valid)))
```

```python
from sklearn import metrics
fpr,tpr,_ = metrics.roc_curve(y_valid,kn.predict_proba(x_valid)[:,1])
plt.plot(fpr,tpr,linestyle = 'solid',color = 'red')
plt.stackplot(fpr,tpr,color = 'steelblue')
plt.plot([0,0],[1,1],linestyle = 'dashed',color = 'black')
plt.text(0.6,0.4,'AUC=%.3f'%metrics.auc(fpr,tpr),fontdict = dict(size = 18))
plt.show()
```

```python
grid_kn_pred = grid_kn.predict(x_valid)
print(pd.crosstab(grid_kn_pred,y_valid))
print('grid_kn model\' accuary on training dataset is {}'.format(grid_kn.score(x_train,y_train)))
print('grid_kn model\' accuary on validing dataset is {}'.format(grid_kn.score(x_valid,y_valid)))

fpr,tpr,_ = metrics.roc_curve(y_valid,grid_kn.predict_proba(x_valid)[:,1])
plt.plot(fpr,tpr,linestyle = 'solid',color = 'red')
plt.stackplot(fpr,tpr,color = 'steelblue')
plt.plot([0,0],[1,1],linestyle = 'dashed',color = 'black')
plt.text(0.6,0.4,'AUC=%.3f'%metrics.auc(fpr,tpr),fontdict = dict(size = 18))
plt.show()
```

```python
gbdt_pred = gbdt.predict(x_valid)
print(pd.crosstab(gbdt_pred,y_valid))
print('gbdt model\' accuary on training dataset is {}'.format(gbdt.score(x_train,y_train)))
print('gbdt model\' accuary on validing dataset is {}'.format(gbdt.score(x_valid,y_valid)))

fpr,tpr,_ = metrics.roc_curve(y_valid,gbdt.predict_proba(x_valid)[:,1])
plt.plot(fpr,tpr,linestyle = 'solid',color = 'red')
plt.stackplot(fpr,tpr,color = 'steelblue')
plt.plot([0,0],[1,1],linestyle = 'dashed',color = 'black')
plt.text(0.6,0.4,'AUC=%.3f'%metrics.auc(fpr,tpr),fontdict = dict(size = 18))
plt.show()
```

```python
grid_gbdt_pred = grid_gbdt.predict(x_valid)
print(pd.crosstab(grid_gbdt_pred,y_valid))
print('grid_gbdt model\' accuary on training dataset is {}'.format(grid_gbdt.score(x_train,y_train)))
print('grid_gbdt model\' accuary on validing dataset is {}'.format(grid_gbdt.score(x_valid,y_valid)))

fpr,tpr,_ = metrics.roc_curve(y_valid,grid_gbdt.predict_proba(x_valid)[:,1])
plt.plot(fpr,tpr,linestyle = 'solid',color = 'red')
plt.stackplot(fpr,tpr,color = 'steelblue')
plt.plot([0,0],[1,1],linestyle = 'dashed',color = 'black')
plt.text(0.6,0.4,'AUC=%.3f'%metrics.auc(fpr,tpr),fontdict = dict(size = 18))
plt.show()
```
